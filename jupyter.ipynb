```python
import pandas as pd
import matplotlib.pyplot as plt
```

1️⃣ Ler o CSV

```python
try:
    tabela = pd.read_csv("tarifas_tusd.csv", sep=";", encoding="utf-8")
except UnicodeDecodeError:
    tabela = pd.read_csv("tarifas_tusd.csv", sep=";", encoding="latin1")
```

2️⃣ Converte coluna VlrTUSD para float

```python
tabela['VlrTUSD'] = tabela['VlrTUSD'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)

display(tabela.head())
```

3️⃣ SubGrupo B4

```python
subgrupo_b4 = tabela[tabela['DscSubGrupo'] == 'B4']
display(subgrupo_b4)
```

4️⃣Filtra a distribuidora ESE (Energisa) para comparção gráfica do aumento da TUSD de 2010 a 2025

```python
ese = subgrupo_b4[subgrupo_b4['SigAgente'] == 'ESE'].copy()

# Converte data para datetime
ese['DatInicioVigencia'] = pd.to_datetime(ese['DatInicioVigencia'], format='%Y-%m-%d')

# Agrupar por ano e calcular média, min e max
ese_por_ano = ese.groupby(ese['DatInicioVigencia'].dt.year)['VlrTUSD'].agg(['mean','min','max'])

# Plotar evolução
plt.figure(figsize=(12,6))
plt.plot(ese_por_ano.index, ese_por_ano['mean'], marker='o', label='Média')
plt.plot(ese_por_ano.index, ese_por_ano['min'], marker='v', linestyle='--', label='Mínimo')
plt.plot(ese_por_ano.index, ese_por_ano['max'], marker='^', linestyle='--', label='Máximo')

plt.xticks(ese_por_ano.index)
plt.title("Evolução da TUSD - ENERGISA (SubGrupo B4)")
plt.xlabel("Ano de Início da Vigência")
plt.ylabel("VlrTUSD (R$)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
